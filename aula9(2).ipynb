```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


seeds_df = pd.read_csv('seeds.csv', header=None)


samples = seeds_df.to_numpy()


samples_train, samples_test = train_test_split(samples, test_size=0.3, random_state=42)


model = KMeans(n_clusters=3, random_state=42)
model.fit(samples_train)


labels = model.predict(samples_train)


x = samples_train[:, 0]  
y = samples_train[:, 2]  

plt.scatter(x, y, c=labels)

plt.show()


```
